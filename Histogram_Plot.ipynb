```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = np.random.normal(loc=34, scale=5, size=2000)
```

```python
plt.hist(x=ages)
```

```python
plt.hist(x=ages, color='black')
```

```python
plt.hist(x=ages, orientation='horizontal')
```

```python
plt.hist(x=ages, histtype='step')
```

```python
plt.hist(x=ages,bins=30)
```
