```python
import matplotlib.pyplot as plt
import random
```

```python
years = [2000+x for x in range(25)]
weights = [60+random.randint(-10, 10) for x in range(25)]
```

```python
plt.title("weights Over 25 years")
plt.ylabel("weights")
plt.xlabel("Years")
plt.plot(years, weights)
```

```python
weights_ticks = list(range(45,75,2))
weights_ticks
```

```python
plt.title("weights Over 25 years", fontsize=20)
plt.ylabel("weights")
plt.xlabel("Years")
plt.yticks(weights_ticks, [f"{x} Kg" for x in weights_ticks])
plt.plot(years, weights)
```
