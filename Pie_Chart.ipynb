```python
import matplotlib.pyplot as plt
import random
```

```python
language = ["python", "c++", "java", "javascript", "dart", "mojo"]
explodes = [0.3,0,0.1,0,0.2,0]
votes = []

for i in range(len(language)) :
    votes.append(random.randint(25,150))
```

```python
plt.pie(x=votes, labels=language)
```

```python
plt.pie(x=votes, labels=language, explode=explodes)
```

```python
plt.pie(x=votes, labels=language, autopct="%.2f%%")
```

```python
plt.pie(x=votes, labels=language,startangle=180)
```
