```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.random_integers(low=0, high=10, size=100)
y = np.random.random_integers(low=0, high=100, size=100)
```

```python
plt.scatter(x=x, y=y)
```

```python
plt.scatter(x=x, y=y, color='brown')
```

```python
plt.scatter(x=x, y=y, marker="*")
```

```python
plt.scatter(x=x, y=y, marker="*", s=200)
```

```python
plt.scatter(x=x, y=y, marker="*", s=150, alpha=0.5)
```
