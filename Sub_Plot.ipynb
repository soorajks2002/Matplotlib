```python
import matplotlib.pyplot as plt
import random
```

```python
x1 = [random.randint(1, 20) for i in range(30)]
x2 = [random.randint(45, 67) for i in range(30)]
```

```python
fig, axis = plt.subplots(2)
```

```python
axis[0].plot(x1)
axis[1].plot(x2)
```

```python
fig
```
